```python
import requests
from bs4 import BeautifulSoup

def scrape_destinypedia_grimoire(url):
    # http get 사용하여 url의 html 파일을 가져옴
    response = requests.get(url)

    # request가 성공했는지 확인
    if response.status_code != 200:
        print("Failed to retrieve the page.")
        return None

    # BeautifulSoup을 사용하여 html 파일을 파싱
    soup = BeautifulSoup(response.content, 'html.parser')

    # html 파일에서 id가 mw-content-text인 div 태그를 찾음
    content_section = soup.find('div', {'id': 'mw-content-text'})

    if content_section is not None:
        # p 태그를 찾음
        paragraphs = content_section.find_all('p')

        data_dict = {}

        # current_verse는 현재 처리 중인 절을 저장, current_text는 현재 처리 중인 절의 텍스트를 저장
        current_verse = None
        current_text = ""

        # enumerate를 사용하여 p 태그를 순회하면서 데이터를 추출
        for idx, paragraph in enumerate(paragraphs):
            # 첫 번째 p 태그는 일반 정보를 담고 있음
            if idx == 0:
                data_dict['General Info'] = paragraph.get_text().strip()
            else:
                # get_text()를 사용하여 p 태그의 텍스트를 가져옴, .strip()을 사용하여 앞뒤 공백을 제거
                paragraph_text = paragraph.get_text().strip()

                # startsWith()를 사용하여 Verse로 시작하는지 확인 (Verse 1:1, Verse 1:2, ...) startsWith() 접두어 확인
                if paragraph_text.startswith("Verse"):
                    # 만약 current_verse가 None이 아니라면, 즉 이전 절이 있다면, data_dict에 이전 절의 정보를 저장
                    if current_verse is not None:
                        data_dict[current_verse] = current_text

                    # 업데이트된 current_verse와 current_text를 저장
                    current_verse = paragraph_text
                    current_text = ""
                else:
                    # 만약 current_verse가 None이 아니라면, 즉 이전 절이 있다면, current_text에 현재 절의 텍스트를 추가
                    current_text += paragraph_text + "\n"

        # 마지막 절의 정보를 data_dict에 저장
        if current_verse is not None:
            data_dict[current_verse] = current_text

        return data_dict
    else:
        print("Books of Sorrow information not found.")
        return None

if __name__ == "__main__":
    url = "https://www.destinypedia.com/Grimoire:Enemies/Books_of_Sorrow"
    data_dict = scrape_destinypedia_grimoire(url)
    if data_dict:
        print(data_dict)
```
